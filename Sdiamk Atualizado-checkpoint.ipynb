```python
from itertools import combinations

#função para calcular o Diametro Steiner de um grafo

def sdiamk(num_v, ord_S, G):
    
    comb = combinations(num_v , ord_S)
    
    l=list(comb)
    dic={}
    
    print("Todas as possíveis árvores S-Steineir com sua respectiva ordem")

    for i in l:
        print i
        st=G.steiner_tree(i)
        print st.size()
        dic[i]= st.size()
        
    l1=dic.values()
    x=max(l1)
    
    print("Diâmetro S-steiner: sdiamk = %d " %x)
    
    print("Familías de Árvore S-Steiner com ordem %d" %x)
    
    for i in dic:
        if dic[i] == x:
            print i
            st1=G.steiner_tree(i)
            st1.show()
        
```

## Caracterização de Grafos com certos $S$-Diâmetro de Steiner($sdiam_{k}$)


### Caracterização Básica 

```python

G=graphs.CompleteGraph(10) #Grafo completo com 10 vértices
G.show()

ord_S= 4 #Ordem do conjunto S
v=G.vertices()

sdiamk(v,ord_S,G)
```

```python
G=graphs.CompleteBipartiteGraph(5, 5) #Grafo completo Bipartido 
G.show()

ord_S= 4 #Ordem do conjunto S
v=G.vertices()

sdiamk(v,ord_S,G)
```

```python
G=graphs.CycleGraph(8)
G.show()

ord_S= 4 #Ordem do conjunto S
v=G.vertices()

sdiamk(v,ord_S,G)
```
