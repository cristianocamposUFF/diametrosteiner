```
from itertools import combinations

#função para calcular o Diametro Steiner de um grafo

def sdiamk(num_v, ord_S, G):
    
    comb = combinations(num_v , ord_S)
    
    l=list(comb)
    dic={}
    
    print("Todas as possíveis árvores S-Steiner com sua respectiva ordem")

    for i in l:
        print i
        st=G.steiner_tree(i)
        print st.size()
        dic[i]= st.size()
        
    l1=dic.values()
    x=max(l1)
    
    print("Diâmetro S-steiner: sdiamk = %d " %x)
    
    
    #print("Familía Árvore S-Steiner com ordem %d" %x)
    
    for i in dic:
        if dic[i] == x:
            #print i
            st1=G.steiner_tree(i)
            
     
    
    print("Árvore S-Steiner com ordem %d" %x)
    st1.show()
```

## Grafos com $3$-Diâmetro de Steiner

__Teorema 1__  Seja $G$ um grafo conexo de ordem $n$. As seguintes afirmações são equivalentes.
                
_(i)_ $sdiam_{3}(G)=2$
                
_(ii)_ $0 \leq \triangle(\overline{G}) \leq 1$
                
_(iii)_ $n-2 \leq \delta(G) \leq n-1$

```
#Exemplo de um grafo satisfazendo o teorema 1.
G=graphs.CompleteGraph(5)
G.delete_edges([(1,0), (2, 3)])
G.show()

n=5
l=G.degree()
print("Sequência de Grau do grafo G:")
print(l)
delta=min(l)

if delta <= n-2 or delta >=n-1:
    print("O grafo satisfaz a condição do teorema 1")
else:
    print("O grafo não satisfaz a condição do teorema 1")

ord_S= 3 #Ordem do conjunto S
v=G.vertices()

sdiamk(v,ord_S,G)    
```
